# 如何观察数据-以 adult 数据集为例

```python
"""数据集预处理

主要是为了 2024-05-08-shap-knn-v1.0.ipynb 准备
如果
"""

import numpy as np
import pandas as pd
from pathlib import Path
```

```python
# 加载数据路径
try:
    file_path = Path(__file__)
    data_path = file_path.joinpath("..", "..", "..", "input", "adult.csv")
except Exception:
    file_path = Path(".")
    data_path = file_path.joinpath("..", "input", "adult.csv")
# 加载原始数据
df = pd.read_csv(data_path, encoding="latin-1")
```

```python
## 查看数据前几行
df.head()
```

```python
## 查看数据后几行
df.tail()
```

```python
# 查看信息
df.info()
```

```python
## 获取所有 object 列
df.select_dtypes("object").columns
```

```python
# 观察数据分布
print("统计数据的个数\n", df['sex'].value_counts())
print("统计数据的占比\n", df[['sex', 'income']].value_counts(normalize=True))
```

```python
## 数据聚合
grouped = df['sex'].groupby([df['race'], df['sex']])
grouped.describe()
```

```python
## 统计非空
df.count()
```

```python
# 统计 np.nan
df.isna().sum()
```

```python
# 统计非 null
df.notnull().sum()
```

```python
df.isnull().sum()
```

```python
# 统计问号
df2 = df.replace('?', np.nan)
df2.isna().sum()
```
