```python
import numpy as np
import matplotlib.pyplot as plt

# 生成随机向量
n = 1000  # 向量的维度
x = np.random.randn(n)  # 生成n维标准正态分布的随机向量

# 画分布图
counts, bins, _ = plt.hist(x, bins=20, alpha=0.7, width=0.2)  # 设置直方图的柱子数量
plt.plot((bins[:-1] + bins[1:]) / 2, counts, '-ro', alpha=0.3)
plt.xlabel('Value')  # x轴标签
plt.ylabel('Frequency')  # y轴标签
plt.title('Distribution of Random Vector')  # 图表标题

# 在每个柱子上显示频率数
for i in range(len(counts)):
    if counts[i] > 0:
        plt.text(bins[i], counts[i], str(counts[i]), ha='center', va='bottom')

plt.show()
```
